```python
from scipy import stats

cols = ['square_feet', 'cleaning_fee','host_listings_count', 'accommodates', 'number_of_reviews', 'review_scores_value', 'calculated_host_listings_count']

for col in df[cols]:
    check = df[col]
    k2, p = stats.normaltest(check)
    alpha = 0.05
    print("p = {:g}".format(p))
    if p < alpha:  # null hypothesis: x comes from a normal distribution
        print("The null hypothesis can be rejected for column " + str(col) + " - nonparametric")
        print('\n')
    else:
        print("The null hypothesis cannot be rejected for column " + str(col) + " - not nonparametric")
        print('\n')
```

I mean that was obvious looking at the pairplot, but yeah...no parametric distributions in this dataframe.

# Neural Network with SKLearn

```python
#whole dataset

from sklearn.neural_network import MLPClassifier
X = df[signif_factors]
y = df['price_night']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit only to the training data
scaler.fit(X_train)

StandardScaler(copy=True, with_mean=True, with_std=True)

# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)


clf = MLPRegressor(solver='lbfgs', alpha=1e-5,
                   hidden_layer_sizes=(5, 2),
                   random_state=101)

clf.fit(X_train, y_train)  
```

```python
from sklearn import metrics

predictions = clf.predict(X_test)
metrics.mean_squared_error(y_test, predictions)
```

```python
from nltk.corpus import stopwords
```

```python
sw = stopwords.words('english')
```

```python
clean_tokens = []

for list in tokens:
    for word in list:
        if word in sw:
            next
        else:
            clean_tokens.append(word)
```

```python
clean_freq = nltk.FreqDist(clean_tokens)
```
